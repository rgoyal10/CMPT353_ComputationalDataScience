```python
import numpy as np
import pandas as pd
from scipy import stats

def get_stats(filename):
    data = pd.read_csv(filename)
    description =  data.describe()
    mean = description.iloc[[1]]
    standard_deviation = description.iloc[[2]]
    minimum = description.iloc[[3]]
    maximum = description.iloc[[7]]
    r = stats.linregress(data['x'], data['y']).rvalue
    return mean, standard_deviation, minimum, maximum, r

file1 = 'data-1.csv'
file2 = 'data-2.csv'
file3 = 'data-3.csv'
file4 = 'data-4.csv'
file5 = 'data-5.csv'
file6 = 'data-6.csv'


mean1, std1, min1, max1, r1 = get_stats(file1)
mean2, std2, min2, max2, r2 = get_stats(file2)    
mean3, std3, min3, max3, r3 = get_stats(file3)    
mean4, std4, min4, max4, r4 = get_stats(file4)    
mean5, std5, min5, max5, r5 = get_stats(file5)    
mean6, std6, min6, max6, r6 = get_stats(file6)    

# description1 = 'The given x and y values are strongly correlated with each other and increase in one could increase the other as correlation coefficient is positive'
# description2 = 'The given x and y values are strongly correlated with each other and increase in one could increase the other as correlation coefficient is positive'
# description3 = 'The given x and y values are strongly correlated with each other and increase in one could increase the other as correlation coefficient is positive'
# description4 = 'The given x and y are very weakly related to each other and increase in one could decrease the other as correlation coefficient is negative'
# description5 = 'The given x and y are very weakly related to each other and increase in one could decrease the other as correlation coefficient is negative'
# description6 = 'The given x and y values are strongly correlated with each other and increase in one could increase the other as correlation coefficient is positive'


summary_data = pd.DataFrame(columns = ['Data','MeanX','MeanY','StdX','StdY','RangeX','RangeY','correlation'])
row1 = {'Data':1,'MeanX':mean1.x[0],'MeanY':mean1.y[0],'StdX':std1.x[0],'StdY':std1.y[0],'RangeX':str([min1.x[0],max1.x[0]]),'RangeY':str([min1.y[0],max1.y[0]]),'correlation':r1}
summary_data = summary_data.append(row1, ignore_index = True)

row2 = {'Data':2,'MeanX':mean2.x[0],'MeanY':mean2.y[0],'StdX':std2.x[0],'StdY':std2.y[0],'RangeX':str([min2.x[0],max2.x[0]]),'RangeY':str([min2.y[0],max2.y[0]]),'correlation':r2}
summary_data = summary_data.append(row2, ignore_index = True)

row3 = {'Data':3,'MeanX':mean3.x[0],'MeanY':mean3.y[0],'StdX':std3.x[0],'StdY':std3.y[0],'RangeX':str([min3.x[0],max3.x[0]]),'RangeY':str([min3.y[0],max3.y[0]]),'correlation':r3}
summary_data = summary_data.append(row3, ignore_index = True)

row4 = {'Data':4,'MeanX':mean4.x[0],'MeanY':mean4.y[0],'StdX':std4.x[0],'StdY':std4.y[0],'RangeX':str([min4.x[0],max4.x[0]]),'RangeY':str([min4.y[0],max4.y[0]]),'correlation':r4}
summary_data = summary_data.append(row4, ignore_index = True)

row5 = {'Data':5,'MeanX':mean5.x[0],'MeanY':mean5.y[0],'StdX':std5.x[0],'StdY':std5.y[0],'RangeX':str([min5.x[0],max5.x[0]]),'RangeY':str([min5.y[0],max5.y[0]]),'correlation':r5}
summary_data = summary_data.append(row5, ignore_index = True)

row6 = {'Data':6,'MeanX':mean6.x[0],'MeanY':mean6.y[0],'StdX':std6.x[0],'StdY':std6.y[0],'RangeX':str([min6.x[0],max6.x[0]]),'RangeY':str([min6.y[0],max6.y[0]]),'correlation':r6}
summary_data = summary_data.append(row6, ignore_index = True)

summary_data =  summary_data.set_index('Data')
summary_data.to_string('summary.txt')

# print(summary_data)




```
